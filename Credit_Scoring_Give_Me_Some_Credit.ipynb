##Give Me Some Credit Dataset
###Contexto:
En las economías de mercado, los bancos desempeñan un rol fundamental al determinar el acceso al financiamiento y las condiciones bajo las cuales este se otorga. Estas decisiones influyen directamente en la capacidad de inversión de individuos y empresas, siendo el crédito un elemento clave para el correcto funcionamiento del sistema económico.

En este contexto, los modelos de credit scoring se utilizan para estimar la probabilidad de incumplimiento de un prestatario. Estos modelos permiten a las instituciones financieras evaluar el riesgo asociado a la concesión de un crédito, facilitando la toma de decisiones informadas.

El presente proyecto se enmarca en la mejora de estos modelos, mediante la predicción de la probabilidad de que un individuo experimente dificultades financieras en un horizonte de dos años. Para ello, se dispone de un conjunto de datos históricos correspondiente a aproximadamente 250,000 prestatarios, el cual incluye variables relacionadas con el comportamiento crediticio, nivel de endeudamiento e ingresos.

El objetivo principal es desarrollar un modelo predictivo que contribuya a una mejor evaluación del riesgo crediticio y que, potencialmente, pueda ser utilizado tanto por entidades financieras como por los propios individuos para optimizar la toma de decisiones financieras.

##Descripcion de Variables
**RevolvingUtilizationOfUnsecuredLines:** Proporción del saldo usado en tarjetas y líneas de crédito sin garantía respecto al límite total disponible.

**age:** Edad del prestatario en años.

**NumberOfTime30-59DaysPastDueNotWorse:** Número de veces que el cliente tuvo retrasos de 30 a 59 días en los últimos 2 años.

**DebtRatio:** Relación entre las deudas mensuales totales y el ingreso bruto mensual.

**MonthlyIncome:** Ingreso mensual del cliente.

**NumberOfOpenCreditLinesAndLoans:** Número total de créditos abiertos, incluyendo préstamos y líneas de crédito.

**NumberOfTimes90DaysLate:** Número de veces que el cliente tuvo retrasos de 90 días o más.

**NumberRealEstateLoansOrLines:** Número de préstamos hipotecarios y créditos relacionados a bienes raíces.

**NumberOfTime60-89DaysPastDueNotWorse:** Número de veces que el cliente tuvo retrasos de 60 a 89 días en los últimos 2 años.
